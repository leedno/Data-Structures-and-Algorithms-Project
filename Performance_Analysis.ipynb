# Performance Analysis of Algorithms

In this section, we analyze the performance of various algorithms based on their execution times across different input sizes. We evaluate the performance of both sorting algorithms and search algorithms, including **Bubble Sort**, **Insertion Sort**, **Merge Sort**, **Selection Sort**, **Heap Sort**, **Quick Sort**, **BFS**, **Binary Search**, and **DFS**.

## Execution Time Analysis




## Sorting Algorithms Performance

The following chart visualizes the performance of sorting algorithms with increasing input size. We will plot only the sorting algorithms to compare them individually.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('performance_results.csv')

# Plot for sorting algorithms (bubble sort, insertion sort, etc.)
plt.figure(figsize=(10, 6))

# Get the unique sorting algorithms and plot them individually
sorting_algorithms = ['Bubble Sort', 'Insertion Sort', 'Merge Sort', 'Selection Sort', 'Heap Sort', 'Quick Sort']

for algo in sorting_algorithms:
    algo_data = df[df['Algorithm'] == algo]
    plt.plot(algo_data['Input Size'], algo_data['Execution Time'], marker='o', label=algo)

plt.title('Sorting Algorithms Performance Analysis')
plt.xlabel('Input Size')
plt.ylabel('Execution Time (seconds)')
plt.legend(title="Sorting Algorithms")
plt.grid(True)
plt.show()
```

### Observations:
- **Bubble Sort** and **Insertion Sort** show significant increases in execution time as input size grows. This is due to their **O(n^2)** time complexity.
- **Merge Sort**, **Heap Sort**, and **Quick Sort** scale better with input size, as they have a better time complexity of **O(n log n)**.
- **Quick Sort** tends to be faster than **Merge Sort** and **Heap Sort** due to its smaller constant factors, though it can be slower in worst-case scenarios.


## Search Algorithms Performance

Now, let's analyze the performance of search algorithms. We include **BFS**, **Binary Search**, and **DFS** in this analysis. These algorithms are evaluated based on their performance when searching through different input sizes.

We will plot these algorithms on individual graphs.


```python
# Plot for BFS, Binary Search, and DFS
search_algorithms = ['BFS', 'Binary Search', 'DFS']

for algo in search_algorithms:
    algo_data = df[df['Algorithm'] == algo]
    plt.plot(algo_data['Input Size'], algo_data['Execution Time'], marker='o', label=algo)
    plt.title(f'{algo} Performance Analysis')
    plt.xlabel('Input Size')
    plt.ylabel('Execution Time (seconds)')
    plt.grid(True)
    plt.legend(title=f"{algo}")
    plt.show()
```

### Observations:
- **Binary Search** shows very efficient performance, even for larger arrays, due to its **O(log n)** time complexity. This makes it extremely fast compared to BFS and DFS.
- **BFS** and **DFS** show more significant increases in execution time as input size grows. This is expected, as these algorithms often need to traverse large portions of the graph or tree, resulting in longer execution times.
  - **BFS** performs better on sparse graphs, but still grows exponentially with the size of the graph.
  - **DFS** is often slower on larger graphs since it explores nodes recursively.


## Average Execution Time for Each Algorithm

We can calculate the average execution time for each algorithm to get an overall picture of how efficient each one is.

```python
# Calculate the average execution time for each algorithm
avg_execution_times = df.groupby('Algorithm')['Execution Time'].mean()

# Print the average execution time for each algorithm
print("Average Execution Times (in seconds):")
print(avg_execution_times)
```

### Observations:
- The **average execution time** gives us a good indication of the general efficiency of each algorithm across the range of input sizes.
  - For sorting, **Quick Sort** and **Merge Sort** are consistently faster than **Bubble Sort** and **Insertion Sort**.
  - For searching, **Binary Search** has the smallest average execution time due to its logarithmic time complexity.


## Detailed Performance Analysis for Each Algorithm

In this section, we analyze the performance of each algorithm in more detail by printing out the execution times for different input sizes.

```python
algorithms = ['Bubble Sort', 'Insertion Sort', 'Merge Sort', 'Selection Sort', 'Heap Sort', 'Quick Sort', 'BFS', 'Binary Search', 'DFS']

for algo in algorithms:
    print(f"\nPerformance Analysis for {algo}:")
    algo_data = df[df['Algorithm'] == algo]
    print(algo_data[['Input Size', 'Execution Time']].to_string(index=False))
```

### Observations:
- This analysis provides a detailed view of how each algorithm behaves with varying input sizes.
- It helps in identifying algorithms that perform well across different ranges of input sizes, and those that struggle as the input grows larger.


## Conclusions

1. **Sorting Algorithms**:
   - **Quick Sort** is the most efficient sorting algorithm for large input sizes.
   - **Bubble Sort** and **Insertion Sort** should be avoided for larger datasets due to their **O(n^2)** complexity.
   - **Merge Sort** and **Heap Sort** provide good alternatives with **O(n log n)** time complexity.

2. **Search Algorithms**:
   - **Binary Search** is extremely fast for large datasets, with logarithmic time complexity.
   - **BFS** and **DFS** are useful for graph traversal but should be chosen carefully depending on the graph structure and problem requirements.

In general, algorithm choice should depend on the problem at hand. For sorting large datasets, prefer algorithms like **Quick Sort** or **Merge Sort**. For searching sorted arrays, **Binary Search** is the most efficient option, whereas **BFS** and **DFS** are useful for graph traversal tasks.
