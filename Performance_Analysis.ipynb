## Execution Time Analysis

### Sorting Algorithms
The following chart visualizes the performance of sorting algorithms with increasing input size. As expected, algorithms like **Quick Sort** should outperform **Bubble Sort** and **Insertion Sort**, which are less efficient for larger inputs due to their higher time complexity.



```python
# Plot for sorting algorithms
plt.figure(figsize=(10, 6))

# Loop through each sorting algorithm and plot its performance
algorithms = df['Algorithm'].unique()

for algo in algorithms:
    algo_data = df[df['Algorithm'] == algo]
    plt.plot(algo_data['Input Size'], algo_data['Execution Time'], marker='o', label=algo)

plt.title('Sorting Algorithms Performance Analysis')
plt.xlabel('Input Size')
plt.ylabel('Execution Time (seconds)')
plt.legend(title="Algorithms")
plt.grid(True)
plt.show()
```

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load the performance data from the CSV file
# Ensure that your CSV file is located in the same folder as your notebook or provide the full path
df = pd.read_csv('performance_results.csv')

# Display the dataframe to check the contents
df.head()

# Plot the execution times for each sorting algorithm by input size
plt.figure(figsize=(10, 6))

# Loop through each sorting algorithm and plot its performance
algorithms = df['Algorithm'].unique()  # Get unique sorting algorithms

# Plot sorting algorithms
for algo in algorithms:
    algo_data = df[df['Algorithm'] == algo]
    plt.plot(algo_data['Input Size'], algo_data['Execution Time'], marker='o', label=algo)

plt.title('Sorting Algorithms Performance Analysis')
plt.xlabel('Input Size')
plt.ylabel('Execution Time (seconds)')
plt.legend(title="Algorithms")
plt.grid(True)
plt.show()

# Performance analysis for each sorting algorithm
algorithms = ['Bubble Sort', 'Insertion Sort', 'Merge Sort', 'Selection Sort', 'Heap Sort', 'Quick Sort']

for algo in algorithms:
    algo_data = df[df['Algorithm'] == algo]
    plt.plot(algo_data['Input Size'], algo_data['Execution Time'], marker='o', label=algo)
    plt.title(f'{algo} Performance Analysis')
    plt.xlabel('Input Size')
    plt.ylabel('Execution Time (seconds)')
    plt.grid(True)
    plt.legend(title=f"{algo}")
    plt.show()

# Plot performance for BFS, Binary Search, and DFS
search_algorithms = ['BFS', 'Binary Search', 'DFS']

for algo in search_algorithms:
    algo_data = df[df['Algorithm'] == algo]
    plt.plot(algo_data['Input Size'], algo_data['Execution Time'], marker='o', label=algo)
    plt.title(f'{algo} Performance Analysis')
    plt.xlabel('Input Size')
    plt.ylabel('Execution Time (seconds)')
    plt.grid(True)
    plt.legend(title=f"{algo}")
    plt.show()

# Calculate the average execution time for each algorithm
avg_execution_times = df.groupby('Algorithm')['Execution Time'].mean()

# Print the average execution time for each algorithm
print("Average Execution Times (in seconds):")
print(avg_execution_times)

# You can also print the overall performance for each algorithm
for algo in search_algorithms + algorithms:
    print(f"\nPerformance Analysis for {algo}:")
    algo_data = df[df['Algorithm'] == algo]
    print(algo_data[['Input Size', 'Execution Time']].to_string(index=False))
```
